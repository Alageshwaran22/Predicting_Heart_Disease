```python
# removing the outliers
df_sem_outlier = df[df["BP"] < 140+1.5*20]
df_sem_outlier = df[df["Cholesterol"] < 280+1.5*(280-213)]
df_sem_outlier = df[df["Max HR"] > 80]
df_sem_outlier = df[df["ST depression"] < 80]
df_sem_outlier = df[df["Number of vessels fluro"] < 2.5]
df_sem_outlier.reset_index
df_sem_outlier
```

```python
df_sem_outlier.reset_index(inplace=True)
df_sem_outlier.info()
```

```python
# creating a new dataframe to perform a PCA analysis on the continous variables
df_pca = df_sem_outlier[["Age","BP","Cholesterol",'Max HR','ST depression','Number of vessels fluro']]
df_pca.info()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

df_std = scaler.fit_transform(df_pca) # standarization of the data to perform PCA analysis
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=5)
pca.fit(df_std)
print("explained variance: ",pca.explained_variance_ratio_)
print(" ")
print("All components contribuite to the variance")
#print(pca.components_)
df_with_pca = pd.DataFrame(pca.fit_transform(df_std))
df_with_pca
```

```python
df_final = df_with_pca

df_final["Sex"] = df_sem_outlier["Sex"]
df_final["Chest pain type"] = df_sem_outlier["Chest pain type"]
df_final["EKG results"] = df_sem_outlier["EKG results"]
df_final["Exercise angina"] = df_sem_outlier["Exercise angina"]
df_final["Slope of ST"] = df_sem_outlier["Slope of ST"]
df_final["Thallium"] = df_sem_outlier["Thallium"]
```

```python
df_sem_outlier.info()
```

```python
df_final[['EKG 1', 'EKG 2', 'EKG 3']] = pd.get_dummies(df_final["EKG results"]) # performing one-hot enconding
df_final.drop("EKG results",axis=1,inplace=True)
df_final.head()
```

```python
X = df_final
y = pd.get_dummies(df_sem_outlier["Heart Disease"]) # separating in features
y = y["Presence"]                                   # and target to train a ML model
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42) # spliting the dataset
```

```python
X_train.columns = X_train.columns.astype(str)
X_test.columns = X_test.columns.astype(str)
```

```python
from sklearn.linear_model import LogisticRegression
import numpy as np

SEED = 42

clf = LogisticRegression(random_state=SEED).fit(X_train, y_train) # traning the model
y_predict = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_predict) # accuracy of the trained model
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_predict) # confusion matrix
print("")
```
